```python
menheart = heart[(heart.sex == 1)]
womanheart = heart[(heart.sex == 0)]
mensum = len(menheart)
womansum = len(womanheart)
plt.figure(figsize=(4,8))
plt.bar([''],[mensum],color='g',label='',width=0.5)
plt.bar([''],[womansum],color='r',label='',width=0.5)
plt.title("    ")
plt.legend()
plt.show()
```

```python
pd.crosstab(heart.sex,heart.target).plot(kind="bar",figsize=(15,6),color=['r','b' ],alpha = 0.5)
plt.title('       o')
plt.xlabel(' (0 = , 1 = )')
plt.xticks(rotation=0)
plt.legend(["no heart disease","heart disease"])
plt.ylabel('')
plt.show()

```

```python
hearttrue = heart[(heart.target == 1)]
heartfalse = heart[(heart.target == 0)]
sumtrue = len(hearttrue)
sumfalse = len(heartfalse)
plt.figure(figsize=(4,8))
plt.title("    ")
plt.bar(['true'],[sumtrue],color="b",alpha=0.5,label="heart disease",width=0.5)
plt.bar(['false'],[sumfalse],color="r",alpha=0.5,label='no heart disease',width=0.5) 
plt.legend()
plt.ylabel('')
plt.show()
```

```python
pain1 = heart[(heart.cp == 1)]
pain2 = heart[(heart.cp == 2)]
pain3 = heart[(heart.cp == 3)] 
nopain = heart[(heart.cp == 0)] 
sumpain1 = len(pain1)
sumpain2 = len(pain2)
sumpain3 = len(pain3)
sumnopain = len(nopain)
```

```python
plt.figure(figsize=(7,8))
plt.bar(['pain 1'],[sumpain1],color="b",alpha=0.7,label="type 1 pain",width=0.5)
plt.bar(['pain 2'],[sumpain2],color="g",alpha=0.3,label='type 2 pain',width=0.5)
plt.bar(['pain 3'],[sumpain3],color="r",alpha=0.8,label='type 3 pain',width=0.5) 
plt.bar(['no pain'],[sumnopain],color="orange",alpha=1.0,label='no pain',width=0.5) 
plt.title("   ")
plt.legend()
plt.ylabel('ROWS')
plt.show()
```

```python
pd.crosstab(heart.cp,heart.target).plot(kind="bar",figsize=(15,6),color=['g','b'],)
plt.title('       ')
plt.xlabel('   , 0 = nopain 1 = pain1 2 = pain2 3 = pain3')
plt.ylabel('')
plt.legend(("no heart disease","heart disease"))
plt.xticks(rotation=1 )
plt.show()
```
