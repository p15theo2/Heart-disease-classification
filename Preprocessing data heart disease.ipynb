```python
heart['sex'][heart['sex'] == 0] = 'female'
heart['sex'][heart['sex'] == 1] = 'male'
```

```python
heart['chest_pain_type'][heart['chest_pain_type'] == 0] = 'typical angina'
heart['chest_pain_type'][heart['chest_pain_type'] == 1] = 'atypical angina'
heart['chest_pain_type'][heart['chest_pain_type'] == 2] = 'non-anginal pain'
heart['chest_pain_type'][heart['chest_pain_type'] == 3] = 'asymptomatic'
```

```python
heart['fasting_blood_sugar'][heart['fasting_blood_sugar'] == 0] = 'sugar_lower than 120mg/ml'
heart['fasting_blood_sugar'][heart['fasting_blood_sugar'] == 1] = 'sugar_greater than 120mg/ml'
```

```python
heart['rest_ecg'][heart['rest_ecg'] == 0] = 'normal'
heart['rest_ecg'][heart['rest_ecg'] == 1] = 'ST-T wave abnormality'
heart['rest_ecg'][heart['rest_ecg'] == 2] = 'left ventricular hypertrophy'
```

```python
heart['exercise_induced_angina'][heart['exercise_induced_angina'] == 0] = 'no'
heart['exercise_induced_angina'][heart['exercise_induced_angina'] == 1] = 'yes'
```

```python
heart['st_slope'][heart['st_slope'] == 0] = 'upsloping'
heart['st_slope'][heart['st_slope'] == 1] = 'flat'
heart['st_slope'][heart['st_slope'] == 2] = 'downsloping'
```

```python
heart['thalassemia'][heart['thalassemia'] == 1] = 'normal'
heart['thalassemia'][heart['thalassemia'] == 2] = 'fixed defect'
heart['thalassemia'][heart['thalassemia'] == 3] = 'reversable defect'
```

```python
heart['sex'] = heart['sex'].astype('object')
heart['chest_pain_type'] = heart['chest_pain_type'].astype('object')
heart['fasting_blood_sugar'] = heart['fasting_blood_sugar'].astype('object')
heart['rest_ecg'] = heart['rest_ecg'].astype('object')
heart['exercise_induced_angina'] = heart['exercise_induced_angina'].astype('object')
heart['st_slope'] = heart['st_slope'].astype('object')
heart['thalassemia'] = heart['thalassemia'].astype('object')
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
data = pd.read_csv("/home/multispiros/python data science/heart disease/dummyheart.csv")
X = data.drop(['target'],axis =1)
y = data['target'] 
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
print(featureScores.nlargest(15,'Score')) 
```

```python
nlargest = featureScores.nlargest(14,'Score')
largest = nlargest['Specs'].values
large = data[largest]
target = data['target']
fsheart = pd.concat([large,target],axis=1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
data = pd.read_csv("/home/multispiros/python data science/heart disease/dummyv2heart.csv")
X = data.drop(['target'],axis =1)
y = data['target']
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
print(featureScores.nlargest(30,'Score')) 
```

```python
nlargest = featureScores.nlargest(13,'Score')
largest = nlargest['Specs'].values
large = data[largest]
target = data['target']
fsheart = pd.concat([large,target],axis=1)
```
